```python
#assigning project paths
project_path = '/content/drive/MyDrive/'+ ProjectFolderName
videofile_path = ['/content/drive/MyDrive/'+ProjectFolderName+'/videos/']
path_config_file = '/content/drive/MyDrive/'+ProjectFolderName+'/config.yaml'

#checking paths are correct
print('NEW project_path: ' + project_path)
print('NEW path_config_file: ' + path_config_file)
print('NEW videofile_path: ' + videofile_path[0])
```

```python
#create a training set using videos defined in config file
deeplabcut.create_training_dataset(path_config_file, net_type='resnet_50', augmenter_type='imgaug')
```

```python
#train network using this user-labeled training set, set iterations to = 100,000, saving checkpoints along the way
deeplabcut.train_network(path_config_file, shuffle=1, displayiters=1000, saveiters=1500, maxiters=75000)
```

```python
#creating evaluation plots for network
%matplotlib notebook
deeplabcut.evaluate_network(path_config_file,plotting=False)
```

```python
#analysing new videos using trained network and save output as csv
deeplabcut.analyze_videos(path_config_file, ['/content/drive/MyDrive/DRP_APP/videos/MPV4_MB11_2_MFinal.mp4'], save_as_csv=True, videotype='.mp4')
```

```python
#filter predictions
deeplabcut.filterpredictions(path_config_file, ['/content/drive/MyDrive/DRP_APP/videos/MPV4_MA1_2_MFinal.mp4'], filtertype='arima', p_bound=0.05, ARdegree=3, MAdegree=1, alpha=0.01)
```

```python
#create videos labelled with predicted body parts
deeplabcut.create_labeled_video(path_config_file, ['/content/drive/MyDrive/DRP_APP/videos/MPV4_MA1_2_MFinal.mp4'], videotype='.mp4', filtered=True, trailpoints=10)
```

```python
#plot trajectories of body parties over course of video
deeplabcut.plot_trajectories(path_config_file,['/content/drive/MyDrive/DRP_APP/videos/MPV4_MA1_2_MFinal.mp4'], videotype='.mp4', filtered=True)
```
